## If Statements with User-Defined Functions

```python
# Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# checking frequency of order_hour_of_day
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

#### Note: Considering we are working with a 24 hour day, I will separate each group into 8. Most orders will include: 10, 11, 14, 15, 13, 12, 16, 9. Average orders will include: 17, 8, 18, 19, 20, 7, 21, 22 and Fewest orders will include: 3, 4, 2, 5, 1, 0, 6, 23 

```python
hour_of_day = []

for value in ords_prods_merged["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        hour_of_day.append("Most orders")
    elif value in [ 17, 8, 18, 19, 20, 7, 21, 22]: 
        hour_of_day.append("Average orders")
    else:
        hour_of_day.append("Fewest orders")
```

```python
ords_prods_merged['busiest_period_of_day'] = hour_of_day
```
