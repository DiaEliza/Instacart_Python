### 6.  Create a spending flag for each user based on the average price across all their orders

1. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
2. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# creating "average_price" column 
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# creating "Spending_flag" for average price < 10 as low spender
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#creating "Spending_flag" for average price >= 10 as High spender
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

### 7. Create order frequency flag that marks the regularity of a user's ordering behavior 

1. If the median of "days_since_prior_order" is higher than 20, then the customer should be labeled as a "Non-frequent customer."
2. If the median is higher than 10 and lower than or equal to 20 , then the customer should be labeled as "Regular customer"
3. If the median is lower than or equalto 10, then the customer should be labeled a "Frequent customer

```python
# creating "median_orders" column 
ords_prods_merge['median_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Creating "order_frequency_flag" for median_orders > 20 as "Non-frequent customer"
ords_prods_merge.loc[ords_prods_merge['median_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Creating "order_frequency_flag" for median orders > 10 & <= 20 as "Regular customer"
ords_prods_merge.loc[(ords_prods_merge['median_orders'] > 10) & (ords_prods_merge['median_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer' 
```

```python
# Creating "order_frequency_flag" for median orders < 10 as "Frequent customer"
ords_prods_merge.loc[ords_prods_merge['median_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# checking "order_frequency_flag" frequency
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
